```python
# num_ecg_paths = ecg_paths.len()
# for i in range(num_ecg_paths):
#     ecg = u.read_matrix(Path("..", ecg_paths[i]))
#     break
ecg = u.read_matrix(Path(ecg_paths[1456]))
```

```python
num_leads = 12

cross_correlations = []

for a_lead in range(num_leads):
    for b_lead in range(a_lead + 1, num_leads, 1):
        c = correlate(
            in1=ecg[:, a_lead],
            in2=ecg[:, b_lead],
            method="direct"
        )
        cross_correlations.append(
            (a_lead, b_lead, c)
        )
        
```

```python
particular_cross_corr_result_index = 15
px.line(
    x=np.arange(cross_correlations[particular_cross_corr_result_index][2].shape[0]),
    y=np.abs(cross_correlations[particular_cross_corr_result_index][2])
)
```
